```python
genre=['Comedy' ,'Horror', 'Biography' ,'Documentary', 'Drama', 'Mystery' ,'Sci-Fi',
 'Romance', 'War' ,'Animation', 'History', 'Family', 'Fantasy' ,'Sport',
 'Adventure' ,'Western', 'Crime' ,'Thriller' ,'Action', 'Music']
language=['English', 'Spanish' ,'French' ,'Italian', 'Hindi', 'Korean' ,'Japanese',
 'German']
country=['USA', 'Canada' ,'Japan', 'South Korea', 'Germany' ,'India', 'France', 'UK']
content=['TV Series' ,'Movie', 'Stand-up Comedy' ,'Limited Series' ,'Documentary']
```

```python
pre=ColumnTransformer(
    transformers=[
        ("cat",OneHotEncoder(handle_unknown="ignore"),
         ["content_type","genre_primary","country_of_origin","language",]),
        ("num", StandardScaler(), 
         ["user_category", "content_warning", "duration_minutes", "release_year"])
    
    ]
)
```

```python
model=Pipeline([
    ("preprocess", pre),
    ("classifier", LogisticRegression(max_iter=1000,class_weight="balanced"))
])
from sklearn.ensemble import RandomForestClassifier
modelR=Pipeline([
    ("preprocess", pre),
    ("classifier", RandomForestClassifier(n_estimators=200, random_state=42))
])
modelG=Pipeline([
    ("preprocess",pre),
    ("classifier",GradientBoostingClassifier( n_estimators=200,      
    learning_rate=0.01,     
    max_depth=4,            
    min_samples_split=5,    
    min_samples_leaf=2))
])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x
```

```python
model.fit(x_train, y_train)
result=model.predict(x_test)
result
```
